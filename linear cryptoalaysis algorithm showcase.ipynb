```python
import numpy as np

s_box = np.array([10, 4, 3, 11, 8, 14, 2, 12, 5, 7, 6, 15, 0, 1, 9, 13], dtype=np.uint8)
n = 16
s_box
```

```python
linear_equations = np.arange(n**2)
linear_equations = linear_equations[
    np.asarray(linear_equations % n, dtype=bool) & (linear_equations >= n)
]  # filter out equations with no variables on each side
linear_equations
```

```python
plaintext_ciphertext_sequences = s_box + np.arange(n) * n
plaintext_ciphertext_sequences
```

```python
approximations = (
    linear_equations.repeat(n).reshape(-1, n) & plaintext_ciphertext_sequences
)
approximations
```

```python
state = np.zeros(approximations.shape, dtype=bool)
for i in range(n):
    state ^= np.asarray((approximations // 2**i) % 2, dtype=bool)

state
```

```python
true_equations = n - state.sum(axis=1)
biases = np.abs(true_equations / n - 0.5)
biases
```

```python
ordering = biases.argsort()[::-1]
linear_equations[ordering], true_equations[ordering]
```
