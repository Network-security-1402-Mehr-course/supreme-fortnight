```python
import numpy as np

s_box = np.array([10, 4, 3, 11, 8, 14, 2, 12, 5, 7, 6, 15, 0, 1, 9, 13], dtype=np.uint8)
n = 16
bit_count = 4
s_box
```

```python
def x(i):
    return (np.arange(n, dtype=np.uint) & (np.ones(n, dtype=np.uint) << i)) >> i


def y(i):
    return (s_box & (np.ones(n, dtype=np.uint) << i)) >> i


# {1, x3, ... , xo, y3, ... , Yo, x3x2, x3x1, ... , x1xo, x3y3, x3y2, ... , xoyo, y3y2, y3y1, ... ,y1Yo}
monomials = [
    np.ones(n, dtype=np.uint),  # 1
    *[x(i) for i in reversed(range(bit_count))],  # x3, ... , x0
    *[y(i) for i in reversed(range(bit_count))],  # y3, ... , y0
    *[
        x(i) * x(j) for i in reversed(range(bit_count)) for j in reversed(range(i))
    ],  # x3x2, x3x1, ... , x1x0
    *[
        x(i) * y(j)
        for i in reversed(range(bit_count))
        for j in reversed(range(bit_count))
    ],  # x3y3, x3y2, ... , x0y0
    *[
        y(i) * y(j) for i in reversed(range(bit_count)) for j in reversed(range(i))
    ],  # y3y2, y3y1, ... , y1y0
]
monomials
```

```python
space_matrix = np.column_stack(monomials)
for i in space_matrix:
    print(i)
```

```python
from sympy import Matrix

null_space = np.array(Matrix(space_matrix).nullspace())
null_space /= np.min(np.abs(null_space)[null_space != 0])
null_space %= 2
null_space = null_space.reshape(-1, len(monomials))
null_space
```
